```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_smoothing(input_image, filter_size=(5, 5)):
    if input_image is None:
        raise ValueError("Input image not found or unable to load.")


    blurred_image = cv2.blur(input_image, filter_size)
    return blurred_image


gray_image_file = '/content/dog_ivp.jpg'
rgb_image_file = '/content/ivp_img.jpg'

gray_img = cv2.imread(gray_image_file, cv2.IMREAD_GRAYSCALE)
rgb_img = cv2.imread(rgb_image_file)

if gray_img is None:
    print(f"Error: Unable to load grayscale image from {gray_image_file}")
if rgb_img is None:
    print(f"Error: Unable to load RGB image from {rgb_image_file}")

if gray_img is not None and rgb_img is not None:

    smoothed_gray_img = apply_smoothing(gray_img)
    smoothed_rgb_img = apply_smoothing(rgb_img)

    plt.subplot(1, 2, 1)
    plt.title('Smoothed Grayscale Image')
    plt.imshow(smoothed_gray_img, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title('Smoothed RGB Image')
    plt.imshow(cv2.cvtColor(smoothed_rgb_img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def to_grayscale(input_image):
    return cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Load the image and convert it to grayscale
original_image = cv2.imread('/content/ivp_img.jpg')
grayscale_img = to_grayscale(original_image)

# Display the grayscale image
plt.imshow(grayscale_img, cmap='gray')
plt.title('Converted Grayscale Image')
plt.show()
```

```python
def display_grayscale_histogram(grayscale_img):
    plt.hist(grayscale_img.ravel(), bins=256, range=[0, 256])
    plt.title('Grayscale Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

display_grayscale_histogram(grayscale_img)
```

```python
def display_color_histogram(original_image):
    colors = ('b', 'g', 'r')
    for i, color in enumerate(colors):
        hist = cv2.calcHist([original_image], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)
        plt.xlim([0, 256])
    plt.title('Color Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

display_color_histogram(original_image)
```
